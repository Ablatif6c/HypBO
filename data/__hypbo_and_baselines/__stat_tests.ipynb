```python
import pandas as pd
import scikit_posthocs as sp
from scipy.stats import friedmanchisquare

# Load your data
file_path = 'regret - Copy.csv'  # Replace with the path to your file
data = pd.read_csv(file_path)

# Drop the first column
data = data.iloc[:, 1:]

# Assuming the first column is an identifier (like function names)
identifier_col = data.columns[0]

# Perform the Friedman Test
friedman_stat, friedman_p_value = friedmanchisquare(
    *[data[method] for method in data.columns if method != identifier_col])

print(f"Friedman Test Statistic: {friedman_stat}, p-value: {friedman_p_value}")

# If the Friedman test is significant, proceed with Conover post-hoc test
if friedman_p_value < 0.05:
    # Reformatting data for posthoc_conover
    # Transforming data from wide format to long format
    data_long = pd.melt(data, id_vars=identifier_col)
    data_long.columns = [identifier_col, 'method', 'value']

    # Performing Conover post-hoc test
    conover_results = sp.posthoc_conover(
        data_long, val_col='value', group_col='method')

    print("Conover Post-hoc Test Results:")
    print(conover_results)

    # Save the results to a CSV file
    conover_results.to_csv('conover_results.csv', index=True)
else:
    print("Friedman test is not significant; post-hoc analysis may not be necessary.")
```
